```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("students performance.csv")  # file ka naam change ho sakta hai
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df['gender'] = df['gender'].map({'male': 0, 'female': 1})
```

```python
sns.boxplot(x='gender', y='math_score', data=df)
plt.title("Math Score by Gender")
plt.show()
```

```python
sns.barplot(x='parental_level_of_education', y='reading_score', data=df)
plt.xticks(rotation=45)
plt.title("Parental Education vs Reading Score")
plt.show()
```

```python
sns.boxplot(x='test_preparation_course', y='writing_score', data=df)
plt.title("Test Preparation Course vs Writing Score")
plt.show()
```

```python
print("✔️ Students who completed test preparation scored higher.")
print("✔️ Parental education seems to impact reading scores.")
print("✔️ Female students performed slightly better in reading and writing.")
```

```python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average_score'] = df['total_score'] / 3
```

```python
sns.histplot(df['average_score'], kde=True)
plt.title("Distribution of Average Score")
plt.show()
```
